### Read in Data


```python
import pandas as pd

sp = pd.read_csv('../data/sp500_raw.csv')
sp.head()
```

```python
sp.columns
```

### Process and Clean Data

```python
sp['date'] = pd.to_datetime(sp['date'])

sp = sp.sort_values('date')

sp.head()
```

```python
sp = sp.dropna(subset=['date', 'close'])
sp.head(), len(sp)
```

### Save the Result

```python
sp.to_csv('../data/sp500_clean.csv', index=False)
```
